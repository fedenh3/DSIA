### Ejercicio 2 \- Dataset simple con Pandas

```
df = pd.DataFrame({
    "estudiante": ["Ana", "Juan", "Carla", "Pedro", "Lucia"],
    "nota": [8, 6, 9, 5, 7]
})

df
```

```
#calculo el promedio de las notas:
df["nota"].mean()
```

```
#muestro la nota máxima:
df["nota"].max()
```

```
#muestro la nota mínima:
df["nota"].min()
```
