### Ejercicio 2 – Condicionales

```
edad = int(input("Ingrese su edad (en números): "))

if edad >= 18:
    print("Eres es mayor de edad")
else:
    print("Eres es menor de edad")
```

### Ejercicio 3 – Listas y bucles

```
lista = [1, 2, 3, 4, 5]

for número in lista:
    print(número)
```

```
suma = 0

for número in lista:
    suma = suma + número

promedio = suma / 5

print("La suma es:", suma)
print("El promedio es:", promedio)
```
